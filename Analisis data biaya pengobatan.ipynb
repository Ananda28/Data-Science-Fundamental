### SIMPLE LINEAR REGRESSION

```python
x = df['age']
y = df['bmi']
```

```python
sns.set()
plt.axis([15, 68, 0, 60])
plt.xticks(fontsize= 15)
plt.yticks(fontsize= 15)
plt.xlabel('age', fontsize= 15)
plt.ylabel('bmi', fontsize= 15)
plt.plot(x, y, 'go')
```

```python
w = 1
x1 = 0
x2 = 60
y1 = x1 * w
y2 = x2 * w
```

```python
sns.set()
plt.axis([15, 68, 0, 60])
plt.xticks(fontsize= 15)
plt.yticks(fontsize= 15)
plt.xlabel('age', fontsize= 15)
plt.ylabel('bmi', fontsize= 15)
plt.plot(x, y, 'go')
plt.plot([x1,x2], [y1,y2], 'r')
plt.show()
```

```python
def predict(x, w):
    return x * w
```

```python
def loss(x, y, w):
    return np.average((predict(x, w)-y) ** 2)
```

```python
def train(x, y, iterations, lr):
    w = 0
    for i in range(iterations):
        current_loss = loss(x, y, w)
        print('Iterasi %4d -> Loss: %.6f' % (i, current_loss))
        
        if loss(x, y , w + lr) < current_loss:
            w += lr
        elif loss(x, y , w - lr) < current_loss:
            w -= lr
        else:
            return w
    raise Exception('Terjadi kesalahan pada iterasi ke %d' % iterations)
```

```python
w = train(
    x,
    y,
    iterations= 10000,
    lr= 0.01
)
```

```python
print('w %.3f' % w)
```

```python
age = 80
print('Prediksi: bila umur = %d, maka bminya = %.2f' %(age, predict(age, w)))
```
