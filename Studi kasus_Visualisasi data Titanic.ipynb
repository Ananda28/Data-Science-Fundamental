```python
import matplotlib.pyplot as plt
import numpy as np
#Untuk mengarahkan ke gambar statis plot
%matplotlib inline
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('titanic') # load data
df
```

```python
# Plot Line
sns.barplot(x = 'who',
            y = 'age',
            data = df)
 
# Show the plot
plt.show()
```

```python
#Heatmaps
corrmat = df.corr()
f, ax = plt.subplots(figsize=(9, 6))
sns.heatmap(corrmat, vmax=.7, square=True)
```

```python
#Kita bisa gunakan argumen annot=True untuk menampilkan korelasi antar atribut. 
#Jika nilai korelasi mendekati 1 maka hubungan antar atribut semakin tinggi
sns.heatmap(corrmat, vmax=.7, annot=True)
```

```python
# Boxen Plot
sns.catplot(x="fare", y="age", data=df, kind="boxen",height=4, aspect=2.7, hue = "who")
```

```python
# Ridge Plot
# Initialize the FacetGrid object
g = sns.FacetGrid(df, row="who", hue="class", aspect=5, height=3)

# # Draw the densities in a few steps
g.map(sns.kdeplot, "age", shade=True, alpha=1, lw=3.5, bw=.2)
g.map(sns.kdeplot, "age", color="w", lw=2, bw=.2)
g.map(plt.axhline, y=0, lw=2)

# # Define and use a simple function to label the plot in axes coordinates
def label(x, color, label):
    ax = plt.gca()
    ax.text(0, .2, label, color=color, ha="left", va="center", transform=ax.transAxes)
g.map(label, "age")

# Set the subplots to overlap
g.fig.subplots_adjust(hspace=-.25)

# # Remove axes details that don't play well with overlap
g.set_titles("")
g.set(yticks=[])
g.despine(bottom=True, left=True)
```

```python
# Group berdasarkan fare
df_fare = df.groupby('sex')['fare'].mean()
df_fare
```

```python
df_fare.plot()
```

```python
df1=df.groupby('who')['age'].sum().to_frame().reset_index()

# Membuat grafik bar
plt.barh(df1['who'],df1['age'],color = ['#B0E0E6','#F0F8FF','#E6E6FA']) 

# Menambahkan Informasi
plt.title('Chart title')
plt.xlabel('X axis title')
plt.ylabel('Y axis title') 

#Tampilkan Plot
plt.show()
```

```python
sns.set_context("paper", font_scale=3, rc={"font.size":8,"axes.labelsize":5})
# Construct iris plot
sns.swarmplot(x="who", y="age", data=df1)

# Show plot
plt.show()
```
