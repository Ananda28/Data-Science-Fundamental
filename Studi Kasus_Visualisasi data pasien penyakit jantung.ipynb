```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
#Untuk mengarahkan ke gambar statis plot
%matplotlib inline 
```

```python
df = pd.read_csv ('heart.csv') # load data
df
```

### Dataset berisi fitur berikut:
   1. age     : (dalam tahun)
   2. sex     : (1 = laki-laki; 0 = perempuan)
   3. cp      : jenis nyeri dada
   4. trestbps: tekanan darah istirahat (dalam mm Hg saat masuk rumah sakit)
   5. chol    : kolestoral serum dalam mg/dl
   6. fbs     : (gula darah puasa > 120 mg/dl) (1 = benar; 0 = salah)
   7. restecg : hasil elektrokardiografi istirahat
   8. thalach : detak jantung maksimum tercapai
   9. exang   : angina yang diinduksi olahraga (1 = ya; 0 = tidak)
   10. oldpeak: depresi ST yang disebabkan oleh olahraga relatif terhadap istirahat
   11. slope  : kemiringan segmen ST latihan puncak
   12. ca     : jumlah pembuluh utama (0-3) yang diwarnai oleh flourosopy
   13. thal   : 3 = biasa; 6 = cacat tetap; 7 = cacat reversibel
   14. target : 1 atau 0

```python
#pie chart
male =len(df[df['sex'] == 1])
female = len(df[df['sex']== 0])

plt.figure(figsize=(8,6))

# Data to plot
labels = 'Male','Female'
sizes = [male,female]
colors = ['yellowgreen', 'skyblue']
explode = (0, 0)  # explode 1st slice
 
# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=90)
plt.title('Percentages of Sex', y=1.12, size=20) 
plt.axis('equal')
plt.show()
```

```python
# Plot Line
sns.barplot(x = 'sex',
            y = 'age',
            data = df)
 
# Show the plot
plt.show()
```

```python
#Heatmaps
corrmat = df.corr()
f, ax = plt.subplots(figsize=(14, 8))
sns.heatmap(corrmat, vmax=.8, square=True)
```

```python
# Boxen Plot
sns.catplot(x="age", y="chol", data=df, kind="boxen",height=4, aspect=2.7, hue = "sex")
```

```python
# Ridge Plot
# Initialize the FacetGrid object
g = sns.FacetGrid(df, row="age", hue="cp", aspect=5, height=3)

# # Draw the densities in a few steps
g.map(sns.kdeplot, "sex", shade=True, alpha=1, lw=3.5, bw=.2)
g.map(sns.kdeplot, "sex", color="w", lw=2, bw=.2)
g.map(plt.axhline, y=0, lw=2)

# # Define and use a simple function to label the plot in axes coordinates
def label(x, color, label):
    ax = plt.gca()
    ax.text(0, .2, label, color=color, ha="left", va="center", transform=ax.transAxes)
g.map(label, "sex")

# Set the subplots to overlap
g.fig.subplots_adjust(hspace=-.20)

# # Remove axes details that don't play well with overlap
g.set_titles("")
g.set(yticks=[])
g.despine(bottom=True, left=True)
```
