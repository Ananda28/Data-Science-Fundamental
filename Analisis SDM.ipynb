```python
# Tampilkan grafik
sns.catplot(x="gender", y="training_hours", hue = "enrolled_university", kind = "bar", data=clean)
```

```python
# Tampilkan grafik
sns.catplot(x="gender", y="training_hours", hue = "enrolled_university", kind = "point", data=clean)
```

```python
sns.catplot(x="gender", y="training_hours", hue="enrolled_university",
            col="education_level", aspect=.9,
            kind="swarm", data=clean);
```

```python
# Histogram
sns.distplot(clean.training_hours)
```

```python
# Hexplot 
sns.jointplot(x=clean.gender, y=clean.training_hours, kind="hex", data = clean)
```

```python
#Heatmaps
corrmat = clean.corr()
f, ax = plt.subplots(figsize=(9, 6))
sns.heatmap(corrmat, vmax=.8, square=True)
```

```python
# Boxen Plot
sns.catplot(x="gender", y="training_hours", data=clean, kind="boxen",height=4, aspect=2.7, hue = "enrolled_university")
```

```python
# Group 
clean_tHours = clean.groupby('gender')['training_hours'].mean()
clean_tHours
```

```python
clean_tHours.plot()
```

```python
X = 'enrollee_id'
Y = 'training_hours'
sns.set()
plt.axis([0, 60, 0, 60])
plt.xticks(fontsize= 15)
plt.yticks(fontsize= 15)
plt.xlabel('enrollee_id', fontsize= 15)
plt.ylabel('training_hours', fontsize= 15)
plt.plot(X, Y, 'gx')
plt.show()
```
